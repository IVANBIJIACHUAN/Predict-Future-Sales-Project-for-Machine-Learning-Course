```python
import matplotlib.pyplot as plt
with open("verbose.txt","r") as f:
    verbose=f.read()
splt=verbose.replace("\n"," ").replace("\t"," ").replace(":"," ").split(" ")
data=[]
for string in splt:
    try:
        data.append(float(string))
    except:
        continue
data.remove(10.0)
train=data[::2]
val=data[1::2]
print(train)

fig,ax=plt.subplots(1,1)
plt.plot(range(len(train)),train,'b--',label="train")
plt.plot(range(len(train)),val,'r--',label="validation")
ax.set_xlabel("steps")
ax.set_ylabel("rmse loss")
ax.legend()
plt.savefig("trainprocess.png")
plt.show()
            
```
